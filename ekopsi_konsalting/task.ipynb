Подготовка данных для регрессии

```python
data_not_null = data.dropna(subset=['Доход (руб.)'])
```

```python
# Выбираем независимые переменные
X = data_not_null[['Отработанные часы', 'Объем производства', 'Разряд']]
y = data_not_null['Доход (руб.)']
```

```python
# Создание и обучение модели
model = LinearRegression()
model.fit(X, y)
```

```python
# Оценка модели
y_pred = model.predict(X)
mae = mean_absolute_error(y, y_pred)
print(f"Средняя абсолютная ошибка модели: {mae}")
```

```python
# Коэффициенты регрессии
print(f"Коэффициенты модели: {model.coef_}")
print(f"Свободный член (intercept): {model.intercept_}")
```

```python
# Заполнение пропущенных значений
data_missing = data[data["Доход (руб.)"].isnull()]
X_missing = data_missing[["Отработанные часы", "Объем производства", "Разряд"]]
data.loc[data["Доход (руб.)"].isnull(), "Доход (руб.)"] = model.predict(X_missing)
```
